```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch

# Load model
print("Loading AI model...")
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def generate_visual_poem(topic):
    # We use descriptive starters to push the AI toward imagery, not dialogue
    starters = [
        f"The {topic} are",
        "Floating through the",
        "A blanket of",
        "Softly they"
    ]

    poem_lines = []
    print(f"--- Poem: {topic} ---\n")

    current_context = f"A beautiful nature poem about {topic}.\n"

    for i, start_phrase in enumerate(starters):
        # We manually feed the model a "starter" for each line
        line_input = current_context + start_phrase
        input_ids = tokenizer.encode(line_input, return_tensors="pt")

        output_ids = model.generate(
            input_ids,
            max_new_tokens=10,
            do_sample=True,
            temperature=0.7, # Lower temperature = less "random" nonsense
            top_p=0.85,
            repetition_penalty=2.0, # High penalty to stop it from repeating the prompt
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.encode("\n")[0]
        )

        full_line = tokenizer.decode(output_ids[0], skip_special_tokens=True)
        # Clean up to keep only the new line
        clean_line = full_line.replace(current_context, "").split("\n")[0]

        print(clean_line)
        poem_lines.append(clean_line)

        # Add the line we just made to the context so it stays on topic
        current_context += clean_line + "\n"

    print("\n--- Done ---")

generate_visual_poem("falling autumn leaves")
```
